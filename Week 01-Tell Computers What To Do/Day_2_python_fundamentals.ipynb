## Q9.Sum of numbers from 1to N

### problem statement:

```python
n= int(input("enter a number:"))
sum=0
for i in range(1, n+1):
  sum = sum + i
print("Sum=",sum)
```

## Q10.Count Even and Odd numbers

### problem statement:
###     take a number N and Count

### *   total even numbers
### *   total odd numbers



```python
n= int(input("enter a number:"))
even=0
odd=0
for i in range(1,n+1):
  if i % 2==0:
    even +=1
  else:
    odd+=1
print("even count:", even)
print("odd coun:", odd)
```
